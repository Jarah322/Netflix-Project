# <b><span style='color:#fcc36d'>Step 12 |</span><span style='color:#ff6200'> </span>Movies watched per month and year</b>


```python
# 

data['Date'] = pd.to_datetime(data['Date'])
data['Year'] = data['Date'].dt.year
data['Month'] = data['Date'].dt.month

# watched per year
yearly_watches = data.groupby('Year').size()
print(yearly_watches)
# watched per month
monthly_watches = data.groupby('Month').size()
print(monthly_watches)
```

```python
# per year distribution
yearly_watches.plot(kind='bar',rot=0)

plt.title('Year watched')
plt.ylabel('No. of Movies')
```

```python
# movie watched by month distribution

data['Month'] = data['Date'].dt.month
data['Month'].value_counts().sort_index().plot(kind='bar',rot=0)
plt.title('Watched by Month')
plt.xlabel('Month')
plt.ylabel('Number of Movies')
plt.show()
```

### Inference

From the graph movies were watched in March, June, July, August, September, November and December.But not in January,February, April, May and October all in 2024. The month of July recorded the highest number of movies watched followed by december and June. **But why is that so?**

# <b><span style='color:#fcc36d'>Step 13 |</span><span style='color:#ff6200'> </span>Using word clouds or frequency analysis to know the frequently appeared words in the dataset</b>


```python
#Are there popular words or themes in Netflix titles over time?
from wordcloud import WordCloud
text = " ".join(data['Title'].dropna())
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Most Common Words in Netflix Titles")
plt.show()
```

### Inferences
From the diagram Limited , Series Glory Episode,Demon Accident appear most and from this I can say series with those keywords are majority hence the titles consist predominantly of series movies.

# <b><span style='color:#fcc36d'>Step 14 |</span><span style='color:#ff6200'> </span>Knowing last visit days in descending order for every movie title to show periods of inactivity</b>


```python
data['Date'] = pd.to_datetime(data['Date'])
data = data.sort_values('Date')
data['Last Visited'] = data['Date'].diff()
visit = data.sort_values('Last Visited', ascending=False)
visit.head(10)
```


<a id="Libraries"></a>
# <b><span style='color:#fcc36d'>Step 15 |</span><span style='color:#ff6200'> </span>viewing the count of days since last visit</b>


```python

visit['Last Visited'].value_counts()
```

### Inference

Key Observations:

The majority of Constant viewing (67) is the highest

There's a sharp drop-off after 0 days, with 26 days viewing at 1 day

Very few cases extend beyond 5 days

There are some rare long-duration outliers (70, 81 days)

#### *Possible Interpretations:*

This could represent time-to-event data (like customer churn, system failures)

Might show response times or processing durations

Could represent how long items remain in some state (like pending orders)

#### *Distribution Characteristics:*

Highly right-skewed distribution

Most events happen immediately or within a couple days
